# Create an IBL simulation for the following situation. 

An agent that selects between two boxes: Box 1 or Box 2. The box has two attributes: the **value** that range from zero to ten and **color**. Similarity for value is computed linearly, as the difference between them normalized by the maximum length of ten. The colors pink and red are considered 70% similar, and all other color pairs are similar only if identical. 

Rewards: The participants get points equal to the value when the color is pink or red. If the box is of any other color, participants receive a fixed reward of 4 points.

Box 1: Value range is from zero to ten, color: {red, pink}
Box 2: Value range is from zero to ten, color: any other color but red or pink

a) How would you define an agent for the above situation?

b) What are the similarity functions for the above task? Please explain the role of partial matching in the above example. 

c) At the end of the simulation of 100 model participants playing the task for 100 trials, which of the box is selected more and why?


A. Agent Denition

```python
from pyibl import Agent

# define similarity function for color
def color_similarity(c1, c2):
    if c1 == c2:
        return 1.0
    elif (c1 == 'red' and c2 == 'pink') or (c1 == 'pink' and c2 == 'red'):
        return 0.7
    else:
        return 0.0

# define similarity function for value
def value_similarity(v1, v2):
    return 1.0 - abs(v1 - v2) / 10.0

# define reward function
def reward(box):
    if box.color in ['red', 'pink']:
        return box.value
    else:
        return 4

# define boxes
box1 = {'value': 10, 'color': 'red'}
box2 = {'value': 5, 'color': 'blue'}

# define agent
a = Agent(default_utility=10, attributes=['color', 'value'])
#a = Agent(attributes=['color', 'value'])
a.similarity(["color"], color_similarity)
a.similarity(["value"], value_similarity)
#a.color_similarity = color_similarity
#a.value_similarity = value_similarity
a.reward_function = reward
a.mismatch_penalty = 1
a.populate([{'color': 'red', 'value': 0}], 10)
a.populate([{'color': 'blue', 'value': 0}], 5)
```

```python
# run simulation
n_trials = 10
n_participants = 10
box_one=[]
box_two=[]

a.trace=True
for i in range(n_participants):
    a.reset()
    box1_selected_count = 0
    box2_selected_count = 0
    for j in range(n_trials):
        selected_box= a.choose([box1, box2])
        a.respond()
        #selected_box = a.respond()
        if selected_box == box2:
            box2_selected_count += 1
        else:
            box1_selected_count += 1

print(f"Box 1 selected {box1_selected_count} times")
print(f"Box 2 selected {box2_selected_count} times")
```
