```python
#importing libraries
from pyibl import Agent
from random import random
import sys
import math
```

```python
#agent and similarity creation
a = Agent(["n", "c", "d"])
a.similarity(["n", "c", "d"], lambda x, y: 1 - abs(x - y))
```

```python
def run_model(trials, mismatch, f):
    a.trace= True
    a.reset()
    a.mismatch_penalty = mismatch
    a.populate([{"n": 0, }], f(0))
    number_correct = 0
    fraction_correct = []
    for t in range(trials):
        options = [ {"n": random()} for _ in range(5) ]
        choice = a.choose(options)
        best = -sys.float_info.max
        best_choice = None
        for o in options:
            v = f(o["n"])
            if o == choice:
                a.respond(v)
            if v > best:
                best = v
                best_choice = o
        if choice == best_choice:
            number_correct += 1
        fraction_correct.append(float(f"{number_correct / (t + 1):.2f}"))
    return fraction_correct
```

```python
def f(x):
   return 5 * (1 - math.pow(x - 0.72, 2))
```

```python
run_model(10, 50, f)
```
