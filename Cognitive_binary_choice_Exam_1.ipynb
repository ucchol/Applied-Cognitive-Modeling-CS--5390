```python
import pyibl
from pyibl import Agent
import time
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
agent = Agent(default_utility=5)
options = ['A','B','C']
choice = agent.choose(options)

def reward(choice):
  if choice == 'A' or 'B':
    r=100
  else:
    r = 50
  return r

trials = 100
participants = 100

for r in range(trials):
  agent.reset()
  for i in range(participants):
    choice = agent.choose(options)
    r =  reward(choice)
    agent.respond(r)
    agent.instances()


```
